# Salary_data.csv

```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.formula.api as smf
```

```python
data=pd.read_csv('https://excelr.s3.nl-ams.scw.cloud/DataScience-Assignments/Assignments/Simple%20Linear%20Regression/Salary_Data.csv')
```

```python
data.head()
```

# EDA and Data Visualization

```python
data.info()
```

```python
sns.distplot(data['YearsExperience'])
```

```python
sns.distplot(data['Salary'])
```

# Correlation

```python
data.corr()
```

```python
sns.regplot(x=data['YearsExperience'],y=data['Salary'])
```

# Model Building

```python
model1 = smf.ols("Salary~np.log(YearsExperience)",data = data).fit()
```

```python
model2= smf.ols("Salary~np.sqrt(YearsExperience)",data = data).fit()
```

```python
model3 = smf.ols("Salary~np.cbrt(YearsExperience)",data = data).fit()
```

```python
model4 = smf.ols("Salary~np.exp(YearsExperience)",data = data).fit()
```

# Model Testing

```python
model1.summary()
```

```python
model2.summary()
```

```python
model3.summary()
```

```python
model4.summary()
```

```python
#Coefficients
model2.params
```

```python
#t and p-Values
print(model2.tvalues, '\n', model2.pvalues)    
```

```python
#predict value for 150,240
```

```python
#R squared values
(model2.rsquared,model2.rsquared_adj)
```

```python
# Manual prediction for say 3 Years Experience
Salary = (25792.200199) + (9449.962321)*(3)
Salary
```

# Predict for new data point

```python
# Automatic Prediction for say 3 & 5 Years Experience 
newdata=pd.Series([3,5])
newdata
```

```python
data_pred=pd.DataFrame(newdata,columns=['YearsExperience'])
data_pred
```

```python
model2.predict(data_pred)
```

# Delivery_time.csv

```python
import pandas as pd
import numpy as np
import seaborn as sns
data = pd.read_csv("https://excelr.s3.nl-ams.scw.cloud/DataScience-Assignments/Assignments/Simple%20Linear%20Regression/delivery_time.csv")
data.head()
```

# EDA and Data Visualization

```python
data.info
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
# Renaming Columns
data1=data.rename({'Delivery Time':'delivery_time', 'Sorting Time':'sorting_time'},axis=1)
data.head()
```

```python
sns.distplot(data1['delivery_time'])
```

```python
sns.distplot(data1['sorting_time'])
```

# Correlation

```python
data1.corr()
```

```python
sns.regplot(x=data1['sorting_time'],y=data1['delivery_time'])
```

# Model Building

```python
import statsmodels.formula.api as smf
model1=smf.ols("delivery_time~np.log(sorting_time)",data = data1).fit()
```

```python
model2=smf.ols("delivery_time~np.sqrt(sorting_time)",data = data1).fit()
```

```python
model3=smf.ols("delivery_time~np.cbrt(sorting_time)",data = data1).fit()
```

```python
model4=smf.ols("delivery_time~np.exp(sorting_time)",data = data1).fit()
```

# Model Testing

```python
model1.summary()
```

```python
model2.summary()
```

```python
model3.summary()
```

```python
model4.summary()
```

```python
model2.params
```

```python
#t and p-Values
print(model2.tvalues, '\n', model2.pvalues)
```

```python
#R squared values
(model2.rsquared,model2.rsquared_adj)
```

```python
# Manual prediction for say sorting time 5
delivery_time = (6.582734) + (1.649020)*(5)
delivery_time
```
